## Informações Estatísticas da Violência no Rio de janeiro

Com os dados adquiridos podemos fazer levantamentos estatísticos para obter melhores *insights*. Podemos obter informações como a média, a mediana e o desvio padrão.

Um método muito usado para se obter uma rápida analise exploratória das variáveis numéricas é o método `deacribe()`. Com esse método, além de obter informações estatísticas, é possível verificar os valores ausentes e outlies.

```
# imprimindo um resumo estatístico do DataFrame
df.describe()
```

Com o método supracitado, conseguimos obter informações tais como:

* cout -  quantidades de entradas validas;
* mean - média dos valores;
* std - desvio padrão;
* min - menor valor da coluna;
* 25% - percentil de 25%;
* 50% - percentil 50% (mediana);
* 75% - percentil 75%;
* max - maior valor da coluna.

Para saber a mpedia de alguma variável podemos analisar o resumo estatístico acima ou usar o método `df.coluna_desejada.mean()`.

Vamos analisar as variáveis `roubo_veiculo`, `furto_veiculos` e `recuperação_veiculos`.

```
# média robo_veiculo
df.roubo_veiculo.mean()
```

```
# média furto_veiculos
df.furto_veiculos.mean()
```

```
# média recuperação_veiculos
df.recuperacao_veiculos.mean()
```

Assim, observa-se que a média de roubo de veiculos é aproximadamente 2447, de furto de veiculos é 1675 e de recuperação de veiculos é de aproximadamente 1772.

Podemos saber ainda, a porcentagem de carros recuperados em relação a quantidade de carros roubados mais a quantidade de carros furtados.

* $\frac{\text{recuperacao_veiculo}}{\text{roubo_veiculo} + \text{furto_veiculos}}$

```
# calculando a porcentagem de carros recuperados
df.recuperacao_veiculos.mean() / (df.roubo_veiculo.mean() + df.furto_veiculos.mean())
```

Assim, conclui-se que a porcentagem de recuperação de veiculos é de aproximadamente 43%.

Ainda observando o resumo estatístico do DataFrama, obtemos que quantidade mínima e máxima da coluna de homicídio doloso, ou seja, quando há a intenção de matar, é de 272 e 831 respectivamente.
